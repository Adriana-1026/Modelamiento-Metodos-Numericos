## **2. HTML Python Library**
Ahora, una imagen cargada con código **HTML** explícito:

```python
from IPython.display import HTML
html1 = '<img src="https://raw.githubusercontent.com/ucmadcursoPythonyR/2020/master/Pics/horse2.jpg" />'
HTML(html1)
```

## **3. Matplotlib Python Library using mpimg (only PNG)**
También se puede usando código **Python** (Se necesita tener instalada la librería **Matplotlib**):

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Without pillow package, matplotlib can only read PNG images
path = 'https://raw.githubusercontent.com/ucmadcursoPythonyR/2020/master/Pics/flow.png';

img = mpimg.imread(path);

plt.imshow(img);
#plt.axis("Off");
plt.show();
```
