```python
from P2Solver_EDOS import *
import numpy as np
import matplotlib.pyplot as plt

α=2/3
β=4/3
δ=1
γ=1


f1=lambda t,x,y:α*x-β*x*y
f2=lambda t,x,y:δ*x*y-γ*y

x0=0.9
y0=0.9
T=[0,200]
n=10**6


t1,x1,y1=RK38(f1,f2,x0,y0,T,n)

x=np.linspace(np.min(x1),np.max(x1),30)
y=np.linspace(np.min(y1),np.max(y1),30)
X,Y=np.meshgrid(x,y)

u=f1(T,X,Y)
v=f2(T,X,Y)
N=np.sqrt(u**2+v**2)
u1=u/N
v1=v/N
plt.figure(figsize=(12,8))
plt.quiver(X,Y,u1,v1)
plt.plot(x1,y1)
plt.grid()
plt.show()
```
