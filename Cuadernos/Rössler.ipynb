```python
from Solver_EDOS import RK38,np # Solver de EDOS
import matplotlib.pyplot as plt # graficar
from mpl_toolkits.mplot3d import Axes3D

#### Hiper Par

# Par. Exp.
A=0.1
B=0.1
C=14.

# Modelo
F=lambda t,X:np.array([-X[1]-X[2],
                      X[0]+A*X[1],
                      B+X[2]*(X[0]-C)])

# Tiempo
T=[0,100]
#Cond. Iniciales
X0=[1,1,1]
# Par Núm.
n=50000

# Sol. Núm
t,X=RK38(F,X0,T,n)
```

```python
plt.figure(figsize=(18,18))

plt.subplot(211)
plt.plot(t,X[:,0],'r',label="x")
plt.plot(t,X[:,1],'g',label="y")
plt.plot(t,X[:,2],'k',label="z")
plt.legend()
plt.grid()

plt.subplot(212,projection="3d")
plt.plot(X[:,0],X[:,1],X[:,2],'-',lw=1.5)
#plt.plot(X[0:30000,0],X[0:30000,1],X[0:30000,2],'k--')
plt.show()
```
