```python
from P3Solver_EDOS import RK4
import matplotlib.pyplot as plt

### P EX

a=40
b=3
c=28

## HP
#Modelo
f1=lambda t,x,y,z:a*(y-x)
f2=lambda t,x,y,z:(c-a)*x-x*z+c*y
f3=lambda t,x,y,z:x*y-b*z

##CI
x0=0
y0=1
z0=0

##Tiempito
T=[0,60]

##PN

n=10**5

# Sol Num
t,x,y,z=RK4(f1,f2,f3,x0,y0,z0,T,n)

# Gráficas
plt.figure(figsize=(15,7))

plt.plot(t,x)
plt.plot(t,y)
plt.plot(t,z)

plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

plt.figure(figsize=(15,10))
plt.subplot(111,projection="3d")
plt.plot(x,y,z,'k.',ms=0.2)
plt.show()
```
