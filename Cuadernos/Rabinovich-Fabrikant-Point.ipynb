```python
from P3Solver_EDOS import RK4
import matplotlib.pyplot as plt

### P EX

γ=0.87
α=1.1

## HP
#Modelo
f1=lambda t,x,y,z:y*(z-1+x**2)+γ*x
f2=lambda t,x,y,z:x*(3*z+1-x**2)+γ*y
f3=lambda t,x,y,z:-2*z*(α+x*y)

##CI
x0=-1
y0=0
z0=0.5

##Tiempito
T=[0,70]

##PN

n=10**5

# Sol Num
t,x,y,z=RK4(f1,f2,f3,x0,y0,z0,T,n)

# Gráficas
plt.figure(figsize=(15,7))

plt.plot(t,x)
plt.plot(t,y)
plt.plot(t,z)

plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

plt.figure(figsize=(15,10))
plt.subplot(111,projection="3d")
plt.plot(x,y,z,'k.',ms=0.03)
plt.show()
```
