```python
from Solver_EDO import *
import numpy as np
import matplotlib.pyplot as plt

f=lambda t,x:np.exp(-t**2)
x0=-1
T=[-3,3]
n=100


t1,x1=RK38(f,x0,T,n)

t=np.linspace(T[0],T[1],30)
x=np.linspace(np.min(x1),np.max(x1),30)
T,X=np.meshgrid(t,x)

u=0.5
v=f(T,X)
N=np.sqrt(u**2+v**2)
u1=u/N
v1=v/N
plt.figure(figsize=(12,8))
plt.quiver(T,X,u1,v1)
plt.plot(t1,x1)
plt.grid()
plt.show()
```
