```python
from Solver_EDOS import RK38,np # Solver de EDOS
import matplotlib.pyplot as plt # graficar

###
g=9.8
c_d=0.25
m=68.1
L=30
k=40
γ=8

## Modelo

F= lambda t,X:np.array([-X[1],
                        g-c_d/m*X[1]**2-(X[0]<-L)*(k/m*(-X[0]-L)+γ/m*X[1])])

F1=lambda t,X:np.array([-X[1],
                        g-c_d/m*X[1]*abs(X[1])-(X[0]<-L)*(k/m*(-X[0]-L)+γ/m*X[1])])

### Con in
X0=[0,0]

#tiempo
T=[0,100]
#par num
n=1000

t,X=RK38(F,X0,T,n)
t,X1=RK38(F1,X0,T,n)
```

```python
plt.figure(figsize=(22,18))

plt.subplot(221)
plt.plot(t,X[:,0],'r',label="Posición x, Incorrect Drag")
plt.plot(t,X1[:,0],'g',label="Posición x, Correct Drag")
plt.legend()
plt.grid()

plt.subplot(222)
plt.plot(t,X[:,1],'r',label="Posición y, Incorrect Drag")
plt.plot(t,X1[:,1],'g',label="Posición y, Correct Drag")
plt.grid()
plt.legend()

plt.subplot(2,2,(3,4))
plt.plot(X[:,0],X[:,1],'r--',label="$x$ Vs $v_x$, Incorrect Drag")
plt.plot(X1[:,0],X1[:,1],'g',label="$x$ Vs $v_x$, Correct Drag")
plt.legend()
plt.grid()

'''
plt.subplot(224)
plt.plot(X[:,2],X[:,3],'r',label="$y$ Vs $v_y$, Incorrect Drag")
plt.plot(X1[:,2],X1[:,3],'g',label="$y$ Vs $v_y$, Correct Drag")
plt.legend()
plt.grid()

plt.figure(figsize=(9,8))
plt.subplot(111)
plt.plot(X[:,0],X[:,2],'r--',label="Trayectoria, Incorrect Drag")
plt.plot(X1[:,0],X1[:,2],'g--',label="Trayectoria, Correct Drag")
#plt.xlim([0,3])
plt.legend()
plt.grid()'''

#plt.subplot(122)
#plt.plot(t,X[:,1],'r',label="Velocidad, Incorrect Drag")
#plt.plot(t,X1[:,1],'g',label="Velocidad, Correct Drag")
#plt.legend()
#plt.grid()
plt.show()
```
